```python
import pandas as pd

# read in the original dataframe
df = pd.read_csv('News_sentiment_Jan2017_to_Apr2021.csv')

# count the number of occurrences of each label
sentiment_counts = df['sentiment'].value_counts()

# determine the minimum count of any label
min_count = sentiment_counts.min()

# create a new dataframe with an equal number of samples for each label to prevent underfitting as well as overfitting
new_df = pd.DataFrame()
for sentiment in sentiment_counts.index:
    sentiment_df = df[df['sentiment'] == sentiment].sample(n=min_count, random_state=42)[:30000]
    new_df = pd.concat([new_df, sentiment_df])

# shuffle the rows of the new dataframe
new_df = new_df.sample(frac=1, random_state=42)[:60000]

# save the new dataframe to a CSV file
new_df.to_csv('Sentiment_news_dataframe.csv', index=False)
```
